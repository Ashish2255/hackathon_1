```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from tqdm.auto import tqdm
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv('Train.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
label_encoder = preprocessing.LabelEncoder()
df['A']= label_encoder.fit_transform(df['A'])
df['A'].unique()
```

```python
df
```

```python
X=df.drop(['Target','A'], axis=1)
y=df['Target']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.1)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=0.001, max_iter=1000, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
y_pred = lasso_reg.predict(X_val)
```

```python
mean_squared_error(y_val, y_pred)
```

```python
y_pred1 = lasso_reg.predict(X_train)
```

```python
mean_squared_error(y_train, y_pred1)
```

```python
Index=test_data['Index']
test_data.drop(['Index','A'],axis=1,inplace=True)
```
